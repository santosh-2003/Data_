# Gender Distribution

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(data = df, x = 'Gender', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

*form the above graph we can see that no. of females in the data set is slightly more than no. of males*

```python
df.columns
```

```python
parents_on_score = df.groupby(['ParentEduc']).agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
parents_on_score
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(data = parents_on_score, cmap="coolwarm", annot = True)
plt.title("Relationship between Parent's education and Student's score", fontsize = 20)
plt.show()
```

*Form the above chart we have concluded that education of prents have good impact on their score* 

# Impact on the score due to Parents marital Status 

```python
marital_status_socre = df.groupby(['ParentMaritalStatus']).agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
sns.heatmap(data = marital_status_socre,  cmap="coolwarm", annot = True)
plt.title("Relationship between Parent's marital status and Student's score", fontsize = 20)
plt.show()
```

*Form the above chart we can observe that marital status of prents have no impact or negligible impact on scores* 

```python
sports_score = df.groupby(['PracticeSport']).agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
sns.heatmap(data = sports_score, cmap = 'coolwarm', annot = True)
plt.title("Relationship between PracticeSport and Student's score", fontsize = 20)
plt.show()
```

# Distribution of Ethnic Groups

```python
print(df['EthnicGroup'].unique())
```

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

group_labels = ['groupA', 'groupB', 'groupC', 'groupD', 'groupE']
mylist = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
plt.title('Distribution of Ethnic Groups', fontsize = 20)
plt.pie(x = mylist, autopct = '%1.2f%%', labels = group_labels)
plt.show()
```
